```python
import csv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from math import *
import numpy as np
import scipy.misc as sm
```

```python
def cropper(x_img,x0,y0):
    x0_end = x0 + 224
    y0_end = y0 + 224
    cropped_img = x_img[x0:x0_end,y0:y0_end]
    #cropped_gt = x_gt[x0:x0_end,y0:y0_end]
    
    return cropped_img#,cropped_gt
```

```python
x_img = mpimg.imread('1PRED.JPG')
plt.imshow(x_img,vmin = 0,vmax =255)
```

```python
for x in range(26):
    for y in range(17):
        
        c_img = cropper(x_img,y*224,x*224)
        temp = 'Images_for_FCRN/PRED/img1/img1_'
        cnt = str(17*x + y + 1)
        po = '.jpg'
        #pgt = 'gt.jpg'
        patho = temp + cnt + po
        #pathgt = temp + cnt + pgt
        sm.imsave(patho,c_img)
        #sm.imsave(pathgt,c_gt)
```
